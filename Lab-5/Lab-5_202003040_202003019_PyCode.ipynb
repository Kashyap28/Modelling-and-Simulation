```python
no_of_obsev=100
time_vals=np.zeros(no_of_obsev)
no_of_suscept_t=np.zeros(no_of_obsev) # no of susceptible at time t
#no_of_suscept[0]=s_t
no_of_infected_t=np.zeros(no_of_obsev) # no of infected at time t
no_of_recover_t=np.zeros(no_of_obsev)  # no of recovered at time t
#total_people_t=np.zeros(no_of_obsev)

cur_time=0
for i in range(no_of_obsev):
    time_vals[i]=cur_time
    no_of_suscept_t[i]=s_t
    no_of_infected_t[i]=i_t
    no_of_recover_t[i]=r_t
    #total_people_t[i]=s_t+r_t+i_t
    
    s_t_plus_del_t=(-1*beta*(s_t/N)*i_t)*delta_t+s_t
    i_t_plus_del_t=((beta*(s_t/N)*i_t)-alpha*i_t)*delta_t+i_t
    r_t_plus_del_t=alpha*i_t*delta_t+r_t
        
    #print(s_t_plus_del_t,i_t_plus_del_t,r_t_plus_del_t)
        
    if(s_t_plus_del_t>0):
        s_t=s_t_plus_del_t
        i_t=i_t_plus_del_t
        r_t=r_t_plus_del_t
    else:
        s_t=0
    cur_time+=delta_t
    
# no_of_suscept_t=N
# no_of_infected_t=N
# no_of_recover_t=N
#total_people_t/=N
    
plt.figure(figsize=(12,7))
str="Plot of S(t),I(t) and R(t) vs t for alpha={},beta={} and R_0={}"
alpha=np.round(alpha,3)
beta=np.round(beta,3)
str=str.format(alpha,beta,np.round(beta/alpha,3))
plt.title(str)
plt.xlabel('Time(in days)')
plt.ylabel('Fraction of people in a particular state')
plt.plot(time_vals,no_of_suscept_t,label='Susceptible at t')
plt.plot(time_vals,no_of_infected_t,label='Infected at t')
plt.plot(time_vals,no_of_recover_t,label='Recover at t')
#plt.plot(time_vals,total_people_t,label='Total people at t')
plt.legend()
ind_max=np.argmax(no_of_infected_t)
print('Day at which max no of infected people found:- ',ind_max/4.0)
print('Max fraction of ppl who get inf at any time t',no_of_infected_t[ind_max])
print('Fraction of people who were infected:- ',no_of_recover_t[-1]) # no of ppl who have recovered at any time t
```
